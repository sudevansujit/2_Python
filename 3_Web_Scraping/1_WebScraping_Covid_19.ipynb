# Web Scraping Covid 19 Data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.worldometers.info/coronavirus/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')

mylist = []
for i in soup.find_all('td'):
    mylist.append(i.text)

Country_Name = mylist[0::10]
Total_Cases = mylist[1::10]
Deaths = mylist[3::10]
Recovered = mylist[5::10]
Active_Cases = mylist[6::10]
Serious_Cases = mylist[7::10]

data = pd.DataFrame(list(zip(Country_Name,Total_Cases,Deaths,Recovered,Active_Cases,Serious_Cases)),
                    columns = ['Country_Name','Total_Cases','Deaths','Recovered','Active_Cases','Serious_Cases'])  
data.head()
```
